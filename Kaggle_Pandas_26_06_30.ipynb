A map is a term, borrowed from mathematicss, for a function that takes one set of values and "maps" them to another set of values. In data science, we often have a need for creating new representations from existing data, or for transforming data from the format it is now in to the format that we want it to be in later. Maps are what handle this work, making them extremely important for getting your work done.


There are two mapping methods that we use often.

````python
#map()
football_height_cm_mean=football.height_cm.mean()
football.height_cm.map(lambda p: p-football_height_cm_mean)

#football.height_cm.mean()을 이용해 선수들 키의 평균을 왼쪽 변수에 저장한다.
#출력된 결과는 height_cm 칼럼을 centering한 series. 해당 칼럼만 반환.
````

map() 함수와 함께 apply() 함수가 자주 사용된다.

apply() is the equivalent method if we want to transform a whole DataFrame by calling a custom method on each row.

````python
a=football.creativity_score.mean()

def remean_creativity_score(row):
  row.creativity_score=row.creativity_score-a
  return row

football.apply(remean_creativity_score, axis='columns')
#이때 football DataFrame 원본은 그대로 저장되어 있다.
#football.head()를 출력하면 원본이 그대로 나온다.
#원본을 바꾸려면 column 전체 변수를 새로 지정해주어야 한다?
````

즉, apply()는 pandas 객체에 사용자가 정의한 함수를 적용할 때 사용된다. Series에 적용할 때는 map()과 같이 lambda x: 구문을 이용하여 각 원소에 함수를 적용한다. 반면 DataFrame에 쓰면 axis 설정에 따라 함수에 들어가는 단위가 달라지는데, axis=0 / axis='index'는 모두 함수를 각 열에 적용한다. 즉, 함수의 입력을 각 열으로 받아 함수를 적용한다. 반면 예시에서처럼 axis=1 / axis='columns'을 쓰면 함수를 각 행에 적용하고, 입력을 한 행 전체로 받는다.

여기서 lambda는 '익명함수'로 부르는 하나의 함수인데, lambda (입력값) : (반환값)의 구조로 작성한다.

Map 함수는 주로 series의 각 원소에 함수를 적용하여 주로 한 열의 값을 다를 값으로 변환하는데 반해, apply 함수는 sereis에도 그리고 DataFrame에도 모두 활용 가능하다. Series에 적용할 때는 map()과 거의 비슷하게 작동하고, DataFrame에 적용할 때는 행 또는 열 단위로 적용하여 여러 행 또는 여러 열을 동시에 참고할 때 유용하다.

````python
def classify_player(row):
  if row.height_cm>=185 and row.creativity_score>=60:
    return "good player"
  elif row.height_cm<=175 and row.creativity_score<=40:
    return "substitute player"
  else:
    return "other"

football["player_type"]=football.apply(classify_player, axis=1)
football

#원하는 칼럼 위치에 "player_type" 칼럼을 넣고 싶다면
#player_type=football.apply(classify_player, axis=1)
#football.insert(n(위치), "player_type"(인덱스 이름), player_type(value)) 처럼 하면 된다
#단, .insert() 메서드는 football DataFrame을 직접적으로 바꿈에 주의.
````

**Lesson 7. Grouping and Sorting**

Maps allow us to transform data in a DataFrame or Series one value at a time for an entire column. However, often we want to group our data, and then do something specific to the group the data is in.

In this section, we study the groupby() operation.

One function we've been using heavily thuus far is the value_counts() function. We can replicate what value_counts() does by doing the following:

````python
football.groupby('creativity_score').creativity_score.count()
#groupby() 안에 넣는 이름이 가져올 칼럼의 index label이 되어야 한다.
#'creativity_score'를 기준으로 groupize하고 각 group에서 creativity_score를 count.
````

We can use any of the summary functions we've used before with this data.

````python
football.groupby('creativity_score').height_cm.min()
#'creativity_score'로 groupize한 카테고리에서 height_cm이 가장 작은 항을 추출한다.
````

````python
football.groupby('creativity_score').height_cm.max()
#'creativity_score'로 groupize한 카테고리에서 height_cm이 가장 큰 항을 추출한다.
````

We can think of each group we generate as being a slice of our DataFrame containing only data with values that match. This DataFrame is accessible to us directly using the apply() method.

````python
football.groupby('team').apply(lambda df: df.player_id.iloc[0])
#football 데이터프레임에서 'team'으로 groupize한 다음, 각 group에서 첫 번째 선수 번호를 뽑아내어 만든 series.
````

We can also group by more than one column. For example, we pick the best creative player by team and age.

````python
football.groupby(["team", "age"]).apply(lambda df: df.loc[df.creativity_score.idxmax()])
#해당선수의 이름만 알고 싶다면 .loc[df.creativity_score(), "player_name"]을 대입.
#먼저 groupby(["team", "age"])에 의해 데이터프레임 전체를 groupize하고
#.apply(lambda 입력 : 출력)을 통해 groupize한 데이터프레임을 하나하나 입력값으로 가져와
#df.creativity_score.idxmax()인 index를 가진 행을 출력하는 코드.
````

Another groupby() method worth mentioning is agg(), which let us run a bunch of different functions on our DataFrame simultaneously.

agg()를 이용하면 여러 함수를 한 번에 데이터프레임에 적용하여 새로운 데이터프레임을 얻을 수 있다.

````python
football.groupby(['nationality']).height_cm.agg([len, min, max])
#len()함수는 그 group에 몇 개의 term이 있는지를 알려준다. 예를들어 Algerian group에는 총 1066명의 선수가 포함되어 있다.
````

**Sorting**

Looking again at countries_age, we can see that grouping returns data in index order, not in value order. That is, when outputting the result of a groupby, the order of the rows is dependent on the values in the index, not int the data. We can use sort_values() method to get data in the order we want.

````python
countries_age=countries_age.reset_index()
countries_age.sort_values(by='len')
#len이라는 데이터값을 기준으로 countreis_age를 sorting하였다.
````

````python
#sort_values의 기본값은 오름차순 / 내림차순으로 sorting하고 싶으면
countries_age.sort_values(by='len', ascending=False)
#ascending=False 코드를 추가하여 len을 내림차순으로 정리하였다.
````

````python
#다시 index 순서대로(원래대로) 배열하고 싶으면
countries_age.sort_index()
````

````python
#한 번에 여러 칼럼을 기준으로 sorting할 수도 있다.
countries_age.sort_values(by=['len', 'age'])
#이러면 len 오름차순 / 나이 오름차순으로 sorting된다.
#어떤 것이 앞에 오느냐에 따라 그것을 기준으로 먼저 sorting하고 두 번째 기준을 다음 sorting 기준으로 반영한다.
````
