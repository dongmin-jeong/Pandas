There are two core objects in pandas: the **DataFrame** and the **Series**. A DataFrame is a table.

```python
pd.DataFrame({'Yes':[50, 21], 'No':[131, 2]})
```

"0, No" entry has the value of 131. "0, Yes" entry has the value of 50.

DataFrame entries are not limited to integers.

```python
pd.DataFrame({'Bob':['I liked it.', 'It was awful.'], 'Sue':['Pretty good.', 'Bland.']})
```

We are using the pd.DataFrame() constructor to generate these DataFrame objects. The syntax is basically a dictionary, whose keys are the column names and values are a list of entries.

To assign the row labels, we use an index parameter in our constructor.

```python
pd.DataFrame({'Bob':['I liked it.', 'It was awful'], 'Sue':['Pretty good.', 'Bland.']}, index=['Product A', 'Product B'])
```

A Series, by contrast, is a sequence of data values.A DataFrame is a table, while a Series is a list. We create nothing more than a list.

```python
pd.Series([1, 2, 3, 4, 5])
```

```python
pd.Series(['a', 'b', 'c'])
```

A Series is, in essence, a single column of a DataFrame. So we can assign row labels to the Series by the same way as before, using an index parameter. Howefer, a Series does not have a column label.

```python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```

The Series and DataFrame are intimately related. It's helpful to think of a DataFrame as actually being just a bunch of Series, glued together.

```python
from google.colab import files
import os
import zipfile
import pandas as pd

uploaded = files.upload()

print("현재 폴더:")
print(os.listdir())

zip_files = [f for f in os.listdir() if f.endswith(".zip")]
csv_files = [f for f in os.listdir() if f.endswith(".csv") and os.path.isfile(f)]

if csv_files:
    print("CSV 파일 발견:", csv_files)
    football = pd.read_csv(csv_files[0])

elif zip_files:
    print("ZIP 파일 발견:", zip_files)
    zip_path = zip_files[0]

    with zipfile.ZipFile(zip_path, "r") as z:
        print("ZIP 내부 파일:")
        print(z.namelist())

        csv_inside = [name for name in z.namelist() if name.endswith(".csv")]
        football = pd.read_csv(z.open(csv_inside[0]))

else:
    raise FileNotFoundError("CSV 또는 ZIP 파일을 찾지 못했습니다.")

football.head()
```

지금까지 iloc operator에 대해 실습하였다. iloc operator는 index-based selection이다. 이와 달리 loc operator는 label-based selection이다. 여기서는 해당하는 key에 맞는 value가 출력된다.

```python
football.loc[54596, 'player_name']
```

```python
football.loc[:, ['player_name', 'nationality', 'creativity_score']]
#loc operator와 :를 이용하면 원하는 column을 뽑아낼 수 있다.
```

```python
football.set_index(["player_name", "nationality"])
#.set_index() 메서드를 이용하면 column을 row label으로 바꿀 수 있다. 예시처럼 여러 개를 만들 수도 있다.
```

```python
football.head()
#중요한 것은, .set_index()를 실행해도 원본은 바뀌지 않는다는 것이다.
#원본을 바꾸려면 football=football.set_index()와 같이 지정해주어야 한다.
```

```python
#Conditional selection
football.nationality=='Spanish'
#bool type으로 모두 판별한다.
```

```python
football.loc[football.team=='South Korea']
#이름 뭔데
#loc operator가 label-based selection이므로 iloc이 아니라 loc을 이용?
```

```python
football.loc[(football.team=='South Korea') & (football.age>=25)]
#&을 이용하여 여러 기준으로 필터링 할 수 있다.
```

```python
football.loc[(football.team=='South Korea') | (football.team=='Japan')]
#파이프 |는 or 연산자.
```

```python
football.loc[football.team.isin(['South Korea', 'Japan', 'Germany'])]
```
