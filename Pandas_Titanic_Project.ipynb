```python
titanic=pd.read_csv('titanic.csv')
titanic.head()#check
```

```python
titanic["Sex_Code"]=titanic.Sex.map(lambda s: 1 if s=="female" else 0)
titanic
```

```python
def age_group(row):
  if row.Age<18:
    return "Child"
  elif 18<=row.Age<60:
    return "Adult"
  elif row.Age>=60:
    return "Senior"

titanic["Age_Group"]=titanic.apply(age_group, axis=1)
titanic
```

```python
A=titanic.groupby(["Sex", "Age_Group"]).Survived.agg(['count', 'sum'])
A["survivor_rate"]=A['sum']/A['count']
A.sort_values("survivor_rate", ascending=False)
#타이타닉호에서 여성과 아이들이 먼저 구조되었다는 말은 대체로 옳다.
```

```python
B=titanic.groupby(["Sex", "Pclass"]).Survived.agg(['count', 'sum'])
B["survivor_rate"]=B['sum']/B['count']
B.sort_values("survivor_rate", ascending=False)
#타이타닉호에서 객실 등급이 높을 수록 생존율이 높은 경향이 있다.
```

```python
def fare_group(row):
  if row.Fare<50:
    return "Very cheap"
  elif 50<=row.Fare<150:
    return "Cheap"
  elif 150<=row.Fare<350:
    return "Moderate"
  elif 350<=row.Fare<450:
    return "Expensive"
  elif row.Fare>=450:
    return "Very expensive"

titanic["Fare_Group"]=titanic.apply(fare_group, axis=1)

T=titanic.groupby(["Sex", "Age_Group", "Fare_Group"]).Survived.agg(['count', 'sum'])
T["survivor_rate"]=T['sum']/T['count']*100
T.sort_values('survivor_rate', ascending=False)
#여성 사망자는 대부분 매우 싼 값의 티켓을 구매한 성인 그리고 아동에서 발생
```
